Define a function to calculate age and zodiac sign from a date of birth and test it.



```python
from datetime import date

def analyze_date_of_birth(dob_str):
    """
    Analyzes a date of birth to calculate age and zodiac sign.

    Args:
        dob_str: Date of birth string in 'YYYY-MM-DD' format.

    Returns:
        A dictionary containing age and zodiac sign.
    """
    # 1. Convert date of birth string to date object
    dob = date.fromisoformat(dob_str)
    today = date.today()

    # 2. Calculate age
    age = today.year - dob.year - ((today.month, today.day) < (dob.month, dob.day))

    # 3. Determine zodiac sign
    month = dob.month
    day = dob.day

    if (month == 3 and day >= 21) or (month == 4 and day <= 19):
        zodiac_sign = 'Aries'
    elif (month == 4 and day >= 20) or (month == 5 and day <= 20):
        zodiac_sign = 'Taurus'
    elif (month == 5 and day >= 21) or (month == 6 and day <= 20):
        zodiac_sign = 'Gemini'
    elif (month == 6 and day >= 21) or (month == 7 and day <= 22):
        zodiac_sign = 'Cancer'
    elif (month == 7 and day >= 23) or (month == 8 and day <= 22):
        zodiac_sign = 'Leo'
    elif (month == 8 and day >= 23) or (month == 9 and day <= 22):
        zodiac_sign = 'Virgo'
    elif (month == 9 and day >= 23) or (month == 10 and day <= 22):
        zodiac_sign = 'Libra'
    elif (month == 10 and day >= 23) or (month == 11 and day <= 21):
        zodiac_sign = 'Scorpio'
    elif (month == 11 and day >= 22) or (month == 12 and day <= 21):
        zodiac_sign = 'Sagittarius'
    elif (month == 12 and day >= 22) or (month == 1 and day <= 19):
        zodiac_sign = 'Capricorn'
    elif (month == 1 and day >= 20) or (month == 2 and day <= 18):
        zodiac_sign = 'Aquarius'
    elif (month == 2 and day >= 19) or (month == 3 and day <= 20):
        zodiac_sign = 'Pisces'
    else:
        zodiac_sign = 'Invalid Date' # Should not happen with valid dates

    # 4. Include a section for other interesting facts (optional)
    # Example: Day of the week born on
    day_of_week = dob.strftime('%A')

    return {'age': age, 'zodiac_sign': zodiac_sign, 'day_of_week_born': day_of_week}

# 6. Test the function with a sample date of birth
sample_dob = '1990-07-15'
results = analyze_date_of_birth(sample_dob)
print(f"Analysis for date of birth {sample_dob}:")
print(f"Age: {results['age']}")
print(f"Zodiac Sign: {results['zodiac_sign']}")
print(f"Day of the week born on: {results['day_of_week_born']}")

sample_dob_2 = '2000-01-01'
results_2 = analyze_date_of_birth(sample_dob_2)
print(f"\nAnalysis for date of birth {sample_dob_2}:")
print(f"Age: {results_2['age']}")
print(f"Zodiac Sign: {results_2['zodiac_sign']}")
print(f"Day of the week born on: {results_2['day_of_week_born']}")
```

Define the shoe size conversion function with error handling and include conversion logic based on common charts.



```python
def convert_shoe_size(size, source_standard, target_standard):
    """
    Converts shoe sizes between different regional standards (e.g., US, UK, EU).

    Args:
        size: The shoe size to convert (numeric).
        source_standard: The standard of the input size (e.g., 'US_Men', 'UK_Women', 'EU').
        target_standard: The standard to convert to.

    Returns:
        The converted shoe size (numeric) or an error message (string).
    """
    # Define conversion factors/offsets based on common charts.
    # Note: Shoe size conversions can vary slightly between brands and charts.
    # These values are approximate and based on commonly found conversion data.
    conversion_chart = {
        'US_Men': {
            'UK_Men': lambda s: s - 0.5,
            'EU': lambda s: s + 33,
            'US_Women': lambda s: s + 1.5
        },
        'UK_Men': {
            'US_Men': lambda s: s + 0.5,
            'EU': lambda s: s + 33.5,
            'US_Women': lambda s: s + 2
        },
        'EU': {
            'US_Men': lambda s: s - 33,
            'UK_Men': lambda s: s - 33.5,
            'US_Women': lambda s: s - 31.5
        },
        'US_Women': {
            'US_Men': lambda s: s - 1.5,
            'UK_Women': lambda s: s - 2,
            'EU': lambda s: s + 31.5
        },
         'UK_Women': {
            'US_Women': lambda s: s + 2,
            'UK_Men': lambda s: s - 2,
            'EU': lambda s: s + 33.5
        }
    }

    # 3. Include error handling for invalid inputs
    if not isinstance(size, (int, float)):
        return "Error: Shoe size must be a number."
    if source_standard not in conversion_chart:
        return f"Error: Unsupported source standard '{source_standard}'. Supported standards are: {list(conversion_chart.keys())}"
    if target_standard not in conversion_chart[source_standard]:
        return f"Error: Unsupported target standard '{target_standard}' for source standard '{source_standard}'. Supported target standards are: {list(conversion_chart[source_standard].keys())}"

    # 2. Implement the conversion logic
    converted_size = conversion_chart[source_standard][target_standard](size)

    return converted_size

# 5. Provide examples of how to use the function
print(f"US Men's 9 to UK Men's: {convert_shoe_size(9, 'US_Men', 'UK_Men')}")
print(f"US Men's 9 to EU: {convert_shoe_size(9, 'US_Men', 'EU')}")
print(f"EU 43 to US Men's: {convert_shoe_size(43, 'EU', 'US_Men')}")
print(f"UK Women's 6 to US Women's: {convert_shoe_size(6, 'UK_Women', 'US_Women')}")
print(f"US Women's 8 to EU: {convert_shoe_size(8, 'US_Women', 'EU')}")
print(f"Invalid size input: {convert_shoe_size('large', 'US_Men', 'UK_Men')}")
print(f"Invalid source standard: {convert_shoe_size(9, 'Invalid_Standard', 'UK_Men')}")
print(f"Invalid target standard: {convert_shoe_size(9, 'US_Men', 'Invalid_Target')}")
```

```python
import random
# This simulation models agents gathering resources from different types of nodes.
# Agents have limited capacity and a goal to gather a certain amount of resources.

class ResourceNode:
    """Represents a source of resources."""
    def __init__(self, resource_type, capacity):
        self.resource_type = resource_type
        self.capacity = capacity
        self.available_resources = capacity

    def gather(self, amount):
        """Attempts to gather a specified amount of resources."""
        if self.available_resources >= amount:
            self.available_resources -= amount
            return amount
        else:
            gathered = self.available_resources
            self.available_resources = 0
            return gathered

    def __str__(self):
        return f"{self.resource_type} Node (Available: {self.available_resources}/{self.capacity})"

class Agent:
    """Represents an agent that gathers resources."""
    def __init__(self, agent_id, capacity):
        self.agent_id = agent_id
        self.capacity = capacity
        self.carried_resources = {}

    def gather_from_node(self, node, amount_to_gather):
        """Agent attempts to gather resources from a node."""
        gathered_amount = node.gather(amount_to_gather)
        if node.resource_type not in self.carried_resources:
            self.carried_resources[node.resource_type] = 0
        self.carried_resources[node.resource_type] += gathered_amount
        print(f"Agent {self.agent_id} gathered {gathered_amount} of {node.resource_type} from {node.resource_type} Node.")

    def report_inventory(self):
        """Reports the resources carried by the agent."""
        print(f"Agent {self.agent_id} Inventory: {self.carried_resources}")

# 2. Python Code Implementation
# Set up the simulation environment
resource_nodes = [
    ResourceNode("Wood", 100),
    ResourceNode("Stone", 150),
    ResourceNode("Wood", 70),
    ResourceNode("Gold", 50)
]

agents = [
    Agent(1, 50),
    Agent(2, 75)
]

# Simulate gathering rounds
print("Starting Resource Gathering Simulation:")
for i in range(3): # Simulate 3 rounds of gathering
    print(f"\n--- Round {i+1} ---")
    for agent in agents:
        # Agents randomly choose a node to gather from
        target_node = random.choice(resource_nodes)
        amount_to_attempt = random.randint(10, 30) # Agents attempt to gather random amounts
        agent.gather_from_node(target_node, amount_to_attempt)

# 4. Display the output
print("\n--- Simulation End ---")
for node in resource_nodes:
    print(node)

print("\nAgent Inventories:")
for agent in agents:
    agent.report_inventory()

# 5. Examples of how to use and interpret output
# The output shows the state of each resource node after the simulation
# (how many resources are left) and the inventory of each agent (what
# resources they collected and in what amounts). This demonstrates a simple
# agent-based simulation with resource management.
```
